```python
!pip install pandas numpy
import pandas as pd
import numpy as np
#concatenation of dataframes
dataFrame1=pd.DataFrame({
    'Units Sold':[8446,3018,1517,3322],
    'Unit Cost':[263.33,90.93,159.42,117.11]
})
dataFrame2=pd.DataFrame({
    'Units Sold':[9845,9528,2844,7299],
    'Unit Cost':[6.92,117.11,117.11,35.84]
})
#vertical concatenation
vertical_concat=pd.concat([dataFrame1,dataFrame2],ignore_index=True)
print("Vertical concatenation:")
print(vertical_concat)
#horizontal concatenation
horizontal_concat=pd.concat([dataFrame1,dataFrame2],axis=1)
print("Horizontal concatenation:")
print(horizontal_concat)
#merging dataframes
dfSE=pd.DataFrame({
    'StudentID':[2,4,6,8,10,12],
    'ScoreSE':[98,93,44,77,69,56]
})
dfML=pd.DataFrame({
    'StudentID':[1,2,3,4,5,6],
    'ScoreML':[39,93,49,44,55,78]
})
#inner join
inner_join=pd.merge(dfSE,dfML,on='StudentID',how='inner')
print("\nInner Join")
print(inner_join)
#left join
left_join=pd.merge(dfSE,dfML,on='StudentID',how='left')
print("\nLeft Join")
print(left_join)
#right join
right_join=pd.merge(dfSE,dfML,on='StudentID',how='right')
print("\nRight Join")
print(right_join)
#outer join
outer_join=pd.merge(dfSE,dfML,on='StudentID',how='outer')
print("\nOuter Join")
print(outer_join)
#adding a new column
sales=pd.DataFrame({
    'Units Sold':[8446,3018,1517,3322],
    'Unit Cost':[263.33,90.93,159.42,117.11]
})
sales['Total Cost']=sales['Units Sold']*sales['Unit Cost']
print("\nSales Data")
print(sales)
#descriptive statistics
print("\nDescriptive Statistics")
print(sales.describe())
#value counts
country_df=pd.DataFrame({
    'Country':['Canada','Libya','Japan','Chad','Armenia']
})
print("\nCountry Counts")
print(country_df['Country'].value_counts())
#hierarchical indexing (stack/unstack)
data=np.arange(15).reshape((3,5))
dframe=pd.DataFrame(
    data,
    index=['Rainfall','Humidity','Wind'],
    columns=['Bergen','Olso','Trondheim','Stavanger','Kristiansand']
)
print("\nOriginal DataFrame:")
print(dframe)
#Stack
stacked=dframe.stack()
print("\nStacked Data:")
print(stacked)
#unstack
unstacked=stacked.unstack()
print("\nUnstacked Data:")
print(unstacked)
#series concatenation with keys
series1=pd.Series([0,111,222,333],
                  index=['zeros','ones','twos','threes'])
series2=pd.Series([444,555,666],
                  index=['fours','fives','sixs'])
frame=pd.concat([series1,series2],keys=['Number1','Number2'])
print("\nHierarchical Series:")
print(frame)
print("\nUnstacked Series:")
print(frame.unstack())
```
